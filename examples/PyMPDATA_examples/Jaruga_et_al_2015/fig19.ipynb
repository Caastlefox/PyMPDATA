```python
import numpy as np
import matplotlib.pyplot as plt

N, M  = 200, 200
dx, dy = 10, 15
K_init = 300

net = np.full((N,M), fill_value=K_init, dtype=float)

xi, yi = np.indices((N, M))
mask = np.sqrt((xi*dx-1000)**2+(yi*dy-260)**2) < 250
net += np.where(mask, 0.5, 0)

plt.imshow(net.T,origin='lower')
plt.colorbar()
show_plot('fig_19')
```
